## шаг. Модель прогнозирования оттока клиентов

### Разбивка данных на обучающую и валидационную выборку

```python
X = data.drop('Churn', axis=1)
y = data['Churn']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
#стандартизируем 
scaler = StandardScaler()
scaler.fit(X_train)
X_train_st = scaler.transform(X_train)
X_test_st = scaler.transform(X_test)
```

### Обучение модели на train-выборке 

```python
#обучение логистической регрессией
lr_model = LogisticRegression(solver = 'liblinear', random_state = 0)
lr_model.fit(X_train_st, y_train)
lr_predictions = lr_model.predict(X_test_st)
lr_probabilities = lr_model.predict_proba(X_test_st)[:,1]
```

```python
#обучение случайным лесом
rf_model = RandomForestClassifier(n_estimators = 100, random_state=0)
rf_model.fit(X_train_st, y_train)
rf_predictions = rf_model.predict(X_test_st) 
```

### Оценка метрик accuracy, precision и recall

```python
#оценим метрики для логистической регрессии
print('Метрики логистической регрессии')

print('accuracy:', accuracy_score(y_test, lr_predictions))
print('precision:', precision_score (y_test, lr_predictions))
print('recall:', recall_score (y_test, lr_predictions))
```

```python
#оценим метрики для случайного леса
print('Метрики случайного леса')

print('accuracy:', accuracy_score(y_test, rf_predictions))
print('precision:', precision_score (y_test, rf_predictions))
print('recall:', recall_score (y_test, rf_predictions))
```

### Выводы

* доля правильных ответов для логистической регрессии на 0,9% больше, чем у случайного леса

* доля правильных ответов среди клиентов, попавших в отток, одинакова для обеих моделей

* реальных клиентов, попавших в отток, модель логистической регресии смогла обнаружить на 3,8% больше, чем модель случайного леса

Важно обратить внимание на клиентов, которые действительно планируют отказаться от услуг фитнес-центра. В этом случае метрика recall станет приоритетной, в этой связи модель логистической регрессии показала себя лучше.

## шаг. Кластеризация клиентов

### Стандартизация данных

```python
scaler = StandardScaler()
X_sc = scaler.fit_transform(X)
```

### Матрица расстояний и дендрограмма

```python
#матрица расстояний
linked = linkage(X_sc, method = 'ward')
```

```python
#визуализируем расстояния на дендрограмме
plt.figure(figsize=(15, 10))  
dendrogram(linked, orientation='top')
plt.title('Иерархическая кластеризация')
plt.show()
```

Образовались 4 кластера, из которых один вышел достаточно большим, чтобы принять решение о его разделении. Таким образом, остановимся на 5 кластерах


### Обучение на основании алгоритма K-Means

```python
#зададим 5 кластеров
km = KMeans(n_clusters=5, random_state=0) 
#сформируем вектор кластеров
labels = km.fit_predict(X_sc)
```

```python
#добавим в датафрейм колонку с номером кластера
data['cluster'] = labels
```

```python
data.groupby('cluster').mean().T
```

Есть заметные различия кластеров по участию в партнерских программах, близости к фитнес-центру, участию в групповых занятиях

### Распределение признаков для кластеров

```python
fig = plt.figure(figsize=(20,30))

for i, column in enumerate(data.drop('Churn', axis=1).columns):
    ax = plt.subplot(6, 3, i +1 )
    if column in ['Age','Avg_additional_charges_total', 'Lifetime', 'Avg_class_frequency_current_month']:
        sns.boxplot(data = data, x = 'cluster', y = column, ax = ax)
    else:
        sns.barplot(data = data, x = 'cluster', y = column, ax = ax)
    plt.title('Распределение признака ' + column)
    plt.xlabel('cluster')
    plt.ylabel(column)
    plt.grid()
plt.show()
```

1 кластер с оттоком в 40%: клиенты живут далеко от фитнес-центра, примерно половина является участниками партнерской программы,  редко участвуют в акции "приведи друга", покупают абонемент на срок меньше года, редко посещают групповые программы, меньше других кластеров тратятся на доп. услуги, у них меньше средняя частота посещений в неделю за все время с начала действия абонемента

3 кластер с оттоком в 38%: клиенты живут рядом с фитнес-центром, редко являются участниками партнерской программы, крайне редко участвуют в акции "приведи друга", приобретают абонемент на короткий срок, не посещают групповые занятия, мало тратятся на доп. услуги, на четвертом месте среди всех кластеров по общей продолжительности посещения фитнес-центра и по частоте посещения

4 кластер с оттоком в 20%: клиенты живут рядом с фитнес-центром, редко являются участниками партнерской программы, крайне редко участвуют в акции "приведи друга", приобретают абонемент на средний срок, посещают групповые занятия, на втором месте по тратам на доп. услуги, на третьем месте среди всех кластеров по общей продолжительности посещения фитнес-центра и на втором по частоте посещения

5 кластер с оттоком в 27%: клиенты в относительной близости к фитнес-центру, примерно половина является участниками партнерской программы, на втором месте по участию в акции "приведи друга" (остальные кластеры участвуют крайне редко), в отличие от всех остальных редко оставляют свой номер телефона при регистрации, достаточно часто тратятся на доп. услуги, вторая группа по общей продолжительности посещения фитнес-центра и третья по частоте занятий
